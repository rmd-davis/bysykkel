```python
import pandas as pd
```

```python
base_url = 'https://data.urbansharing.com/trondheimbysykkel.no/trips/v1/2021/'

months = ['01','02','03','04','05','06','07','08','09','10','11']
for month in months:
    url = base_url + month + '.csv'
    if month == '01':
        df = pd.read_csv(url)
    else:
        df_temp = pd.read_csv(url)
        df = pd.concat([df,df_temp])
```

```python
df
```

```python
df.to_excel('data/bysykkel_2021.xlsx')
```

```python
start_stations = df[['start_station_id','start_station_name']].drop_duplicates()
start_stations.to_excel('data/start_station_id.xlsx', index=False)
```
